```python
from gtts import gTTS
from IPython.display import Audio
import time

def speaker(stmt):
    kor_wav = gTTS(stmt, lang = 'ko')
    kor_wav.save('kor.wav')
    display(Audio('kor.wav', autoplay=True))


#1. 단어등록
def w_add(dictionary):
    key=input('새 단어 : ')
    print(f'{key}',end='')
    speaker(f'새로 등록하신 단어는 {key}입니다.')
    value=input('의 뜻 : ')
    speaker(f'새로 등록하신 단어의 뜻은 {value}입니다.')
    dictionary[key]=value

#2. 단어검색
def w_search(dictionary):

    key=input('검색할 단어 : ')
    print(dictionary[key])


#3. 단어삭제
def w_delete(dictionary):
    key = input('삭제할 단어 : ')
    if key in dictionary:  # 딕셔너리에 입력된 키 값이 존재하는지 확인
        del dictionary[key]  # 키 값이 존재하면 해당 단어를 삭제
        print(f'"{key}" 단어 삭제됨!')
    else:
        print(f'"{key}" 단어는 미등록 단어')

#4. 모든단어보기
def w_show(dictionary):
    print(dictionary)

dictionary=dict()    # 빈 딕셔너리 생성

#0. 종료하기
def quit():
    print('프로그램을 종료합니다.')

while True:


    menu='''---선택메뉴--
    1. 단어 등록
    2. 단어 검색
    3. 단어 삭제
    4. 모든 단어 보기
    0. 종료하기
    --------------------'''
    print(menu,end='')

    time.sleep(5)

    speaker(menu)

    time.sleep(2)

    select=input('선택:')

    if select=='1':
        speaker('새 단어를 등록하세요')
        w_add(dictionary)

    elif select=='2':
       speaker('단어를 검색하세요')
       w_search(dictionary)

    elif select=='3':
       speaker('단어를 삭제하세요')
       w_delete(dictionary)

    elif select=='4':
       speaker('모든 단어를 보여주세요')
       w_show(dictionary)

    elif  select=='0':
        speaker('종료합니다.')
        quit()
        break
```

```python
from gtts import gTTS
import pygame


pygame.mixer.init(devicename='pulse')


# pygame 초기화
pygame.mixer.init()


def speaker(stmt):
    kor_wav = gTTS(stmt, lang='ko')
    kor_wav.save('kor.wav')
    pygame.mixer.music.load("kor.wav")
    pygame.mixer.music.play()
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

#1. 단어등록
def w_add(dictionary):
    key=input('새 단어 : ')
    print(f'{key}',end='')
    value=input('의 뜻 : ')
    dictionary[key]=value

#2. 단어검색
def w_search(dictionary):
    key=input('검색할 단어 : ')
    print(dictionary.get(key, '미등록 단어'))

#3. 단어삭제
def w_delete(dictionary):
    key = input('삭제할 단어 : ')
    if key in dictionary:
        del dictionary[key]
        print(f'"{key}" 단어 삭제됨!')
    else:
        print(f'"{key}" 단어는 미등록 단어')

#4. 모든단어보기
def w_show(dictionary):
    print(dictionary)

dictionary = {}  # 빈 딕셔너리 생성

#0. 종료하기
def quit():
    print('프로그램을 종료합니다.')

# 메인 루프
while True:
    menu = '''---선택메뉴--
    1. 단어 등록
    2. 단어 검색
    3. 단어 삭제
    4. 모든 단어 보기
    0. 종료하기
    --------------------'''
    print(menu)
    speaker(menu)
    select = input('선택:')

    if select == '1':
        speaker('새 단어를 등록하세요')
        w_add(dictionary)

    elif select == '2':
        speaker('단어를 검색하세요')
        w_search(dictionary)

    elif select == '3':
        speaker('단어를 삭제하세요')
        w_delete(dictionary)

    elif select == '4':
        speaker('모든 단어를 보여주세요')
        w_show(dictionary)

    elif select == '0':
        speaker('종료합니다.')
        quit()
        break
```
