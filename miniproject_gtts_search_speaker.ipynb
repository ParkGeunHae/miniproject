```python
from gtts import gTTS
from IPython.display import Audio, display
import time

def speaker(stmt):
    kor_wav = gTTS(stmt, lang='ko')
    kor_wav.save('kor.wav')
    display(Audio('kor.wav', autoplay=True))

#1. 단어등록
def w_add(dictionary):
    key = input('새 단어 : ')
    print(f'{key}', end='')
    value = input('의 뜻 : ')
    print(f'등록단어:{key},뜻:{value}')
    speaker(f'새로 등록하신 단어는 {key}이고, 뜻은 {value}입니다.')
    dictionary[key] = value

#2. 단어검색
def w_search(dictionary):
    search_term = input('검색할 단어 또는 뜻 : ')
    found = False

    # key 검색
    if search_term in dictionary:
        meaning = dictionary[search_term]
        print(f'검색하신 단어는 {search_term}이고 뜻은 {meaning}입니다.')
        speaker(f'검색하신 단어는 {search_term}이고 뜻은 {meaning}입니다.')
        found = True

    # value 검색
    value_searched = False  # value 검색 여부를 나타내는 변수 추가
    for key, value in dictionary.items():
        if value == search_term:
            print(f'검색하신 뜻은 {value}이고 단어는 {key}입니다.')
            speaker(f'검색하신 뜻은 {value}이고 단어는 {key}입니다.')
            found = True
            value_searched = True  # value 검색이 실행된 경우 변수를 True로 설정

    # key 검색이나 value 검색이 모두 실행되지 않았고 value 검색이 실행된 적도 없는 경우에만 일부 검색 수행
    if not found and not value_searched:
        for key, value in dictionary.items():
            if search_term in key or search_term in value:
                print(f'검색하신 일부를 포함하는 단어는 {key}이고 뜻은 {value}입니다.')
                # 검색 결과 출력 후 잠시 대기
                speaker(f'검색하신 일부를 포함하는 단어는 {key}이고 뜻은 {value}입니다.')
                time.sleep(5)
                found = True

    # 검색 결과가 없는 경우
    if not found:
        print(f'"{search_term}"에 대한 검색 결과가 없습니다.')
        speaker(f'"{search_term}"에 대한 검색 결과가 없습니다.')

#3. 단어삭제
def w_delete(dictionary):
    key = input('삭제할 단어 : ')
    if key in dictionary:  # 딕셔너리에 입력된 키 값이 존재하는지 확인
        del dictionary[key]  # 키 값이 존재하면 해당 단어를 삭제
        print(f'"{key}" 단어 삭제됨!')
        speaker(f'삭제하신 단어는 {key}입니다.')
    else:
        print(f'"{key}" 단어는 미등록 단어')
        speaker(f'{key}는 미등록 단어입니다.')

#4. 모든단어보기
def w_show(dictionary):
    print(dictionary)
    time.sleep(3)
    speaker(f'등록되어 있는 모든 단어는 {dictionary}입니다.')

dictionary = dict()    # 빈 딕셔너리 생성

#0. 종료하기
def quit():
    print('프로그램을 종료합니다.')

# 메뉴 출력 여부를 나타내는 변수
menu_displayed = False

while True:
    menu = '''---선택메뉴--
    1. 단어 등록
    2. 단어 검색
    3. 단어 삭제
    4. 모든 단어 보기
    0. 종료하기
    --------------------
    원하시는 메뉴를 선택해주세요'''
    print(menu, end='')

    time.sleep(5)

    if not menu_displayed:
        speaker(menu)
        menu_displayed = True

    time.sleep(2)

    select = input('선택:')

    if select == '1':
        speaker('새 단어를 등록하세요')
        w_add(dictionary)

    elif select == '2':
        speaker('단어를 검색하세요')
        w_search(dictionary)

    elif select == '3':
        speaker('삭제할 단어를 입력하세요')
        w_delete(dictionary)

    elif select == '4':
        speaker('모든 단어를 보여주세요')
        w_show(dictionary)

    elif select == '0':
        speaker('종료합니다.')
        quit()
        break
```
